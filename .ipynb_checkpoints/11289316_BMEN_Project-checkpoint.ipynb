### __Abstract:__
This report presents an exploratory analysis of a modified version of the Red Wine dataset using an object-oriented approach in a Jupyter notebook. The primary objective is to gain insights into the characteristics of red wines through descriptive statistics and visualizations. The analysis utilizes the Pandas, NumPy, and Matplotlib libraries to manipulate and visualize the data.


### __Introduction:__
Red wine is a complex beverage with various chemical components that contribute to its taste and quality. This analysis aims to provide a comprehensive understanding of the dataset, exploring key attributes such as fixed acidity,volatile acidity, citric acid, residual sugar, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphate, alcohol, and quality. By employing an object-oriented approach, the analysis is structured and modular, facilitating a systematic exploration of the data.

### __Methodology:__
1. __Data Loading:__
The dataset, a modified version of the Red Wine dataset, was loaded into a Pandas DataFrame. The object-oriented structure ensures encapsulation and ease of data manipulation.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
class WineDataSetAnalysis:
    def __init__(self, WineDataSet):
        self.WineDataSet = pd.read_csv(WineDataSet)
        
    def display_head(self):
        return self.WineDataSet.head(10)

    def display_tail(self):
         return self.WineDataSet.tail(20)

    def display_info(self):
        return self.WineDataSet.info()


    def describeData(self):
        return (self.WineDataSet).describe()


wine_data_analysis = WineDataSetAnalysis('red_wine.csv')
```

2. __Data Overview:__
Basic information about the dataset, including data types and missing values,was displayed using the info method. The first 10 rows and the last 20 rows of the dataset were also presented to provide initial glimpse of the data.

 * ###### INFORMATION ON DATASET

```python
wine_data_analysis.WineDataSet['fixed acidity'] = wine_data_analysis.WineDataSet['fixed acidity'].fillna(8.318335)
wine_data_analysis.WineDataSet['volatile acidity'] = wine_data_analysis.WineDataSet['volatile acidity'].fillna(0.527747)
wine_data_analysis.WineDataSet['chlorides'] = wine_data_analysis.WineDataSet['chlorides'].fillna(0.087476)
wine_data_analysis.WineDataSet['free sulfur dioxide'] = wine_data_analysis.WineDataSet['free sulfur dioxide'].fillna(15.867334)
wine_data_analysis.WineDataSet['density'] = wine_data_analysis.WineDataSet['density'].fillna(0.996746)
wine_data_analysis.WineDataSet['quality'] = wine_data_analysis.WineDataSet['quality'].fillna(5.635795)
```

4. __Data Visualization:__
Matplotlib was employed for data visualization. Histograms were created to illustrate the distribution of key attributes, offering insights into the frequency of specific values. The visualizations provide a clearer understanding of the data distribution.


```python
class WineDataSetAnalysis:
    def __init__(self, WineDataSet):
        self.WineDataSet = pd.read_csv(WineDataSet)


    def plot_histogram(self, column):
        plt.hist(self.WineDataSet[column], bins=6, edgecolor='black', log=True, color= 'green')
        plt.title(f'Histogram of {column}')
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.show()
wine_data_analysis = WineDataSetAnalysis('red_wine.csv')             
wine_data_analysis.plot_histogram('volatile acidity')
```

```python
wine_data_analysis.plot_histogram('chlorides')
```

```python
wine_data_analysis.plot_histogram('alcohol')
```

5. __Correlation Analysis:__
Correlation analysis was conducted to explore relationships between different attributes. A heatmap was generated to visualize the correlation matrix, aiding in the identification of potential patterns and dependencies.


```python
mean = summary.loc['mean']
class WineDataSetAnalysis:
    def __init__(self, WineDataSet):
        self.WineDataSet = pd.read_csv(WineDataSet)

    def plot_correlation_heatmap(self):
        '''This method displays the heatmap of the dataset'''
        corr_matrix = self.WineDataSet.corr()
        plt.figure(figsize=(10, 8))
        plt.imshow(corr_matrix, cmap='coolwarm', interpolation='none')
        plt.colorbar()
        plt.xticks(range(len(corr_matrix)), corr_matrix.columns, rotation='vertical')
        plt.yticks(range(len(corr_matrix)), corr_matrix.columns)
        plt.title('Correlation Heatmap')
        plt.show()


    def draw_pi_chart(self):
         plt.figure(figsize=(10, 10))
         plt.pie(mean, labels=mean.index, autopct='%1.1f%%', startangle=140)
         plt.axis('equal')
         plt.title('Pie Chart(mean)')
         plt.show()



wine_data_analysis = WineDataSetAnalysis('red_wine.csv') 
wine_data_analysis.plot_correlation_heatmap()
```

```python
wine_data_analysis.draw_pi_chart()
```
